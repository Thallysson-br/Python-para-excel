##Joins

```python
#importando bibliotecas
import pandas as pd
from IPython.display import display

#Abrindo Planilhas

clientes_df = pd.read_excel('Join.xlsx')
vendas_df = pd.read_excel('tabela_vendas.xlsx')
vendas_df_join = pd.read_excel('vendas_join.xlsx')


display(clientes_df)
```

```python
display(vendas_df)
```

```python
#Descrição da função
#.merge(Tabela Objeto para mergir com, how= tipo de merge, 
#  left_on= em qual ponto/coluna da tabela original, right_on=em qual ponto/coluna da tabela objeto)
```

```python
#por default a função join usa o inner
incluir_exemplo = True
```

```python
#INNER JOIN
clientes_df.merge(vendas_df_join, how='inner', left_on='Id Compra', right_on='Id Compra')
```

```python
#Left join
clientes_df.merge(vendas_df, how='left', left_on='Id Compra', right_on='Id Compra')
```

```python
#Right join 
clientes_df.merge(vendas_df, how='right', left_on='Id Compra', right_on='Id Compra')
```

```python
#Cross join
clientes_df.merge(vendas_df_join, how='cross')
```

```python
#Outer
clientes_df.merge(vendas_df_join, how='outer', left_on='Id Compra', right_on='Id Compra')
```
